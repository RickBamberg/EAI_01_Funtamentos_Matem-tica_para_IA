### 2. Multiplica√ß√£o de Matrizes

A multiplica√ß√£o de matrizes √© uma opera√ß√£o central na √Ålgebra Linear. Em IA, ela aparece em quase todos os c√°lculos envolvendo redes neurais, transforma√ß√µes lineares e representa√ß√£o de dados.

#### Regras principais:
Para multiplicar duas matrizes:
- A matriz **A** deve ter dimens√µes `m √ó n`
- A matriz **B** deve ter dimens√µes `n √ó p`
- O resultado ser√° uma nova matriz **C** com dimens√µes `m √ó p`

Ou seja, o **n√∫mero de colunas da primeira matriz deve ser igual ao n√∫mero de linhas da segunda**.

#### Exemplo pr√°tico
Vamos visualizar a multiplica√ß√£o de duas matrizes simples:


### üí° Multiplica√ß√£o de Matrizes: Exemplo 2x3 ‚ãÖ 3x2

Se temos:

**Matriz A (2√ó3):**

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
$$

**Matriz B (3√ó2):**

$$
B =
\begin{bmatrix}
7 & 8 \\
9 & 10 \\
11 & 12
\end{bmatrix}
$$

Ent√£o o produto \( C = A \cdot B \) ser√° uma **matriz 2√ó2**, calculada assim:

$$
C = 
\begin{bmatrix}
(1\cdot7 + 2\cdot9 + 3\cdot11) & (1\cdot8 + 2\cdot10 + 3\cdot12) \\
(4\cdot7 + 5\cdot9 + 6\cdot11) & (4\cdot8 + 5\cdot10 + 6\cdot12)
\end{bmatrix}
$$

Fazendo as contas:

- \( C_{11} = 7 + 18 + 33 = 58 \)
- \( C_{12} = 8 + 20 + 36 = 64 \)
- \( C_{21} = 28 + 45 + 66 = 139 \)
- \( C_{22} = 32 + 50 + 72 = 154 \)

**Resultado final:**

$$
C = 
\begin{bmatrix}
58 & 64 \\
139 & 154
\end{bmatrix}
$$


```python
# Matriz A (2x3)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Matriz B (3x2)
B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Multiplica√ß√£o
C = A @ B
print("Resultado da multiplica√ß√£o:")
print(C)
```

#### Interpreta√ß√£o
Cada elemento da matriz resultado √© o produto escalar entre uma linha da primeira matriz e uma coluna da segunda matriz.

Essa opera√ß√£o aparece diretamente em redes neurais, quando multiplicamos as entradas de um neur√¥nio pelos seus respectivos pesos para obter uma ativa√ß√£o.


### üí° Produto de Matrizes ‚Äì F√≥rmula

Queremos multiplicar duas matrizes \( A \) e \( B \).  
Se:

$$
A = \begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}, \quad
B = \begin{bmatrix}
b_{11} & b_{12} \\
b_{21} & b_{22}
\end{bmatrix}
$$

O produto \( C = A \cdot B \) ser√°:

$$
C = \begin{bmatrix}
a_{11}b_{11} + a_{12}b_{21} & a_{11}b_{12} + a_{12}b_{22} \\
a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22}
\end{bmatrix}
$$

---

### ‚úÖ Exemplo pr√°tico

Multiplicando as matrizes:

$$
A = \begin{bmatrix}

\end{bmatrix}, \quad
B = \begin{bmatrix}

$$

O resultado √©:

```python
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

produto_escalar = np.dot(u, v)
print(f"Produto escalar: {produto_escalar}")
```

### üìò Multiplica√ß√£o de Matrizes

Considere as matrizes:

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
\end{bmatrix},
\quad
B =
\begin{bmatrix}
7 & 8 \\
9 & 10 \\
11 & 12 \\
\end{bmatrix}
$$

Multiplicando \( A \cdot B \), obtemos a matriz \( C \):

$$
C =
\begin{bmatrix}
1 \cdot 7 + 2 \cdot 9 + 3 \cdot 11 & 1 \cdot 8 + 2 \cdot 10 + 3 \cdot 12 \\
4 \cdot 7 + 5 \cdot 9 + 6 \cdot 11 & 4 \cdot 8 + 5 \cdot 10 + 6 \cdot 12 \\
\end{bmatrix}
=
\begin{bmatrix}
58 & 64 \\
139 & 154 \\
\end{bmatrix}
$$


```python
# Matrizes
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Multiplica√ß√£o
C = A @ B
print("Resultado da multiplica√ß√£o (C = A @ B):")
print(C)

# Visualiza√ß√£o
fig, ax = plt.subplots(figsize=(6, 6))

# Vetores da base padr√£o
base = np.array([[1, 0], [0, 1]])

# Aplicando transforma√ß√£o da matriz C
for v in base:
    v_transformado = C @ v
    ax.quiver(0, 0, v[0], v[1], color='blue', angles='xy', scale_units='xy', scale=1, label='Base original' if np.array_equal(v, [1, 0]) else "")
    ax.quiver(0, 0, v_transformado[0], v_transformado[1], color='red', angles='xy', scale_units='xy', scale=1, label='Transformado' if np.array_equal(v, [1, 0]) else "")

ax.set_xlim(-10, 160)
ax.set_ylim(-10, 160)
ax.set_aspect('equal')
ax.axhline(0, color='gray')
ax.axvline(0, color='gray')
ax.grid(True)
ax.legend()
ax.set_title("Transforma√ß√£o Linear pela Matriz C = A @ B")
plt.show()
```

### üß† Produto Escalar ‚Äì Intui√ß√£o e Conceito
O produto escalar (ou produto interno) entre dois vetores √© uma opera√ß√£o que gera um n√∫mero (escalar), e √© uma das formas de medir a rela√ß√£o entre eles.

Antes de expandirmos para opera√ß√µes mais avan√ßadas com matrizes, √© importante entender um conceito fundamental: o **produto escalar**.

## üåü O que √© o Produto Escalar?

O produto escalar (ou produto interno) entre dois vetores resulta em um **n√∫mero escalar** (n√£o um vetor). Ele nos ajuda a entender:
* Qu√£o "alinhados" dois vetores est√£o
* O √¢ngulo entre eles
* A proje√ß√£o de um vetor sobre outro

## üìê F√≥rmula Matem√°tica

Se temos dois vetores:

$$\vec{u} = [u_1, u_2, \dots, u_n], \quad \vec{v} = [v_1, v_2, \dots, v_n]$$

O produto escalar √©:

$$\vec{u} \cdot \vec{v} = u_1v_1 + u_2v_2 + \dots + u_nv_n$$

## üßÆ Exemplo Num√©rico

Seja:

$$\vec{u} = [1, 2, 3], \quad \vec{v} = [4, 5, 6]$$

$$\vec{u} \cdot \vec{v} = 1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6 = 4 + 10 + 18 = 32$$

## üîç Interpreta√ß√£o Geom√©trica

O produto escalar tamb√©m pode ser visto como:

$$\vec{u} \cdot \vec{v} = \|\vec{u}\| \cdot \|\vec{v}\| \cdot \cos(\theta)$$

Onde:
* $\|\vec{u}\|$ e $\|\vec{v}\|$ s√£o os m√≥dulos dos vetores
* $\theta$ √© o **√¢ngulo entre os vetores**

## üìê Interpreta√ß√£o Geom√©trica

O produto escalar mede **quanto um vetor "aponta" na dire√ß√£o do outro**. Algumas consequ√™ncias:

* Se o √¢ngulo $\theta = 0¬∞$ ‚Üí $\cos(\theta) = 1$: vetores apontam na **mesma dire√ß√£o**, e o escalar √© **m√°ximo**.
* Se $\theta = 90¬∞$ ‚Üí $\cos(\theta) = 0$: vetores s√£o **ortogonais (perpendiculares)**, e o produto escalar √© **zero**.
* Se $\theta = 180¬∞$ ‚Üí $\cos(\theta) = -1$: vetores em **dire√ß√µes opostas**, e o escalar √© **negativo**.

### üßÆ Exemplo Num√©rico

```python
u = np.array([3, 4])
v = np.array([2, -1])

produto_escalar = np.dot(u, v)
print("Produto escalar:", produto_escalar)
```

### üìä Visualiza√ß√£o

```python
def plot_vectors(u, v):
    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v')
    plt.xlim(-1, 5)
    plt.ylim(-3, 5)
    plt.axhline(0, color='gray')
    plt.axvline(0, color='gray')
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Visualiza√ß√£o dos vetores u e v")
    plt.show()

plot_vectors(u, v)
```

### ‚úÖ Conclus√µes Importantes
Produto escalar √© uma ferramenta essencial para identificar √¢ngulos e rela√ß√µes entre vetores.

Se o produto escalar for zero, os vetores s√£o perpendiculares.

Ele ser√° positivo quando vetores estiverem "na mesma dire√ß√£o", e negativo quando "em dire√ß√µes opostas".

## üî≤ Matrizes Identidade e Inversa

### üß± Matriz Identidade

A matriz identidade $I$ √© o **"1" da multiplica√ß√£o de matrizes**. Ou seja:

$$A \cdot I = I \cdot A = A$$

Para uma matriz 2x2:

$$I = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$$

Ela **n√£o altera** a matriz com a qual √© multiplicada.

### üîÑ Matriz Inversa

A matriz inversa $A^{-1}$ √© aquela que, quando multiplicada por $A$, resulta na matriz identidade:

$$A \cdot A^{-1} = A^{-1} \cdot A = I$$


### üßÆ Exemplo em Python

```python
A = np.array([[2, 1],
              [5, 3]])

# Verificando a identidade
I = np.eye(2)

# Inversa de A
A_inv = np.linalg.inv(A)

# Testando
produto = A @ A_inv

print("Matriz A:")
print(A)

print("\nInversa de A:")
print(A_inv)

print("\nA @ A_inv (deve ser identidade):")
print(produto)
```

### üìê Geometria da Inversa
Visualmente, aplicar uma matriz e depois sua inversa √© como voltar ao ponto original.

Por exemplo: uma transforma√ß√£o que gira um vetor em 90¬∞ e depois aplica a inversa, desfaz a rota√ß√£o.

### ‚ö†Ô∏è Determinante

Para saber se uma matriz tem inversa, usamos o **determinante**.

* Se $\text{det}(A) \neq 0$: matriz invert√≠vel
* Se $\text{det}(A) = 0$: **matriz singular** (sem inversa)

### üßÆ Exemplo em Python

```python
np.linalg.det(A)
```

### ‚õî Casos sem Inversa

```python
B = np.array([[1, 2],
              [2, 4]])

print("Determinante de B:", np.linalg.det(B))
```

Resultado: 0 ‚Üí A matriz √© singular (invers√£o proibida).

## üìä Sistemas Lineares e Solu√ß√£o com Matrizes

### üß© O Problema

Imagine um sistema de equa√ß√µes lineares:

$$\begin{cases} 
2x + y = 5 \\ 
4x - 3y = 6 
\end{cases}$$

Esse sistema pode ser representado como **uma equa√ß√£o matricial**:

$$A \cdot \vec{x} = \vec{b}$$

Onde:
* $A = \begin{bmatrix} 2 & 1 \\ 4 & -3 \end{bmatrix}$
* $\vec{x} = \begin{bmatrix} x \\ y \end{bmatrix}$
* $\vec{b} = \begin{bmatrix} 5 \\ 6 \end{bmatrix}$

### üéØ Objetivo

Resolver o sistema significa encontrar $\vec{x}$, ou seja:

$$\vec{x} = A^{-1} \cdot \vec{b}$$

### üßÆ Solu√ß√£o em Python

```python
# Matriz dos coeficientes
A = np.array([[2, 1],
              [4, -3]])

# Vetor dos resultados
b = np.array([5, 6])

# Solu√ß√£o do sistema
x = np.linalg.solve(A, b)

print("Solu√ß√£o do sistema (valores de x e y):")
print(x)
```

### üìâ Interpreta√ß√£o Geom√©trica
Cada equa√ß√£o representa uma reta no plano. A solu√ß√£o do sistema √© o ponto de interse√ß√£o dessas duas retas.

```python
x_vals = np.linspace(-10, 10, 100)
y1 = (5 - 2 * x_vals)
y2 = (4 * x_vals - 6) / 3

plt.plot(x_vals, y1, label='2x + y = 5')
plt.plot(x_vals, y2, label='4x - 3y = 6')
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.title('Solu√ß√£o de um Sistema Linear: Interse√ß√£o das retas')
plt.show()
```

### üöß Quando n√£o h√° solu√ß√£o?

Existem 3 possibilidades para um sistema:

| Tipo de sistema | Situa√ß√£o geom√©trica | Solu√ß√£o |
|-----------------|---------------------|---------|
| Determinado | Retas se cruzam | 1 solu√ß√£o (interse√ß√£o) |
| Indeterminado | Retas coincidem | ‚àû solu√ß√µes |
| Imposs√≠vel | Retas paralelas | Nenhuma solu√ß√£o |

## üîç Autovalores e Autovetores: Conceito Intuitivo

### üß† Ideia Principal

Imagine que uma transforma√ß√£o linear (representada por uma matriz $A$) muda a dire√ß√£o da maioria dos vetores. Mas existem vetores **especiais** que **mant√™m a sua dire√ß√£o** ap√≥s a transforma√ß√£o ‚Äî apenas mudam de tamanho.

Esses vetores s√£o chamados **autovetores**, e o fator pelo qual s√£o multiplicados √© o **autovalor**.

### üîé Defini√ß√£o Matem√°tica

Dado um vetor $\vec{v} \neq \vec{0}$ e uma matriz quadrada $A$, dizemos que:

$$A \vec{v} = \lambda \vec{v}$$

Se:
* $\vec{v}$: autovetor
* $\lambda$: autovalor correspondente

Isso significa: **multiplicar $A$ pelo vetor $\vec{v}$ √© o mesmo que multiplicar esse vetor por um escalar $\lambda$**.

### üßÆ Exemplo em Python

```python
# Matriz de transforma√ß√£o
A = np.array([
    [2, 1],
    [1, 2]
])

# C√°lculo de autovalores e autovetores
valores, vetores = np.linalg.eig(A)

print("Autovalores:")
print(valores)

print("\nAutovetores (colunas):")
print(vetores)
```

### üéØ Interpreta√ß√£o Geom√©trica
Vamos visualizar como autovetores s√£o transformados pela matriz:

```python
# Pega os dois autovetores
v1 = vetores[:, 0]
v2 = vetores[:, 1]

# Aplica a transforma√ß√£o
Av1 = A @ v1
Av2 = A @ v2

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, Av1[0], Av1[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='A*v1')

#plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
#plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='A*v2')

#plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='v2')
#plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=0.8, color='orange', alpha=0.8, label='A*v2')


plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='v2')
plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=0.8, color='orange', alpha=0.9, label='A*v2')
plt.legend()


plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Autovetores mant√™m a dire√ß√£o ap√≥s transforma√ß√£o")
plt.show()
```

üî∑ Determinante de uma Matriz ‚Äì Significado Geom√©trico e Intui√ß√£o
O determinante de uma matriz quadrada representa, geometricamente, o fator de escala da √°rea (em 2D) ou do volume (em 3D) ap√≥s uma transforma√ß√£o linear.

‚úÖ Intui√ß√£o Visual:
- Se a matriz representa uma transforma√ß√£o no plano (2D), o determinante indica quanto a √°rea √© multiplicada.

- Se o determinante for zero, significa que a transforma√ß√£o achata o espa√ßo, ou seja, colapsa o plano em uma linha (ou ponto).

- Se for negativo, al√©m da escala, h√° uma invers√£o de orienta√ß√£o (reflex√£o).

### üî¢ C√≥digo Python: Visualiza√ß√£o do Efeito do Determinante

```python
def plotar_determinante(matriz, titulo=""):
    # Define o quadrado unit√°rio
    quadrado = np.array([
        [0, 1, 1, 0, 0],
        [0, 0, 1, 1, 0]
    ])

    # Aplica a transforma√ß√£o
    transformado = matriz @ quadrado

    # Calcula o determinante
    det = np.linalg.det(matriz)

    # Plot
    plt.figure(figsize=(6, 6))
    plt.plot(quadrado[0], quadrado[1], 'gray', label='Original')
    plt.plot(transformado[0], transformado[1], 'r', label='Transformado')
    plt.fill(transformado[0], transformado[1], 'r', alpha=0.2)

    plt.title(f"{titulo}\nDeterminante: {det:.2f}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.show()

# Exemplos:
plotar_determinante(np.array([[2, 0], [0, 1]]), "Escala em X (aumenta a √°rea)")
plotar_determinante(np.array([[1, 0], [0, -1]]), "Reflex√£o no eixo Y (inverte a orienta√ß√£o)")
plotar_determinante(np.array([[1, 1], [1, 1]]), "Cisalhamento colapsante (det = 0)")
```

### üìå Conclus√£o:
Determinante = 1 ‚Üí a √°rea n√£o muda.

Determinante > 1 ‚Üí a √°rea aumenta.

Determinante < 0 ‚Üí a √°rea muda e h√° invers√£o de sentido (reflex√£o).

Determinante = 0 ‚Üí colapso de dimens√£o, transforma√ß√£o n√£o-invers√≠vel.

## üî∑ Matriz Inversa ‚Äì Intui√ß√£o, Propriedades e Visualiza√ß√£o

A **matriz inversa** √© o equivalente de "desfazer" uma transforma√ß√£o linear.

Se uma matriz **A** transforma vetores, a **inversa de A**, denotada $A^{-1}$, **desfaz** essa transforma√ß√£o.

### ‚úÖ Intui√ß√£o:

Se aplicarmos uma matriz $A$ e depois sua inversa $A^{-1}$, voltamos ao ponto de origem:

$$A^{-1} \cdot (A \cdot \vec{v}) = \vec{v}$$

Ou seja, **transformar e depois reverter**.

### üß† Condi√ß√£o:

Uma matriz **s√≥ tem inversa se seu determinante for diferente de zero**:

$$\text{Se } \det(A) \neq 0 \Rightarrow A \text{ √© invert√≠vel}$$

### üî¢ Exemplo Num√©rico:
Vamos ver um exemplo simples de invers√£o de matriz 2√ó2.

```python
# Matriz A
A = np.array([
    [2, 1],
    [5, 3]
])

# Inversa de A
A_inv = np.linalg.inv(A)

# Produto A . A‚Åª¬π (deve dar a identidade)
I = A @ A_inv

print("Matriz A:")
print(A)

print("\nMatriz Inversa de A:")
print(A_inv)

print("\nA . A‚Åª¬π (deve ser a identidade):")
print(I)
```

### üéØ Visualiza√ß√£o: Transformar e "Desfazer"

```python
def plotar_inversa(A):
    A_inv = np.linalg.inv(A)

    # Vetores originais
    vetor1 = np.array([1, 1])
    vetor2 = np.array([-1, 1])

    # Aplica a transforma√ß√£o A
    v1_t = A @ vetor1
    v2_t = A @ vetor2

    # Aplica a inversa em seguida
    v1_inv = A_inv @ v1_t
    v2_inv = A_inv @ v2_t

    plt.figure(figsize=(6, 6))

    # Vetores originais
    #plt.quiver(0, 0, vetor1[0], vetor1[1], color='gray', scale=1, scale_units='xy', angles='xy', label='Original v1')
    #plt.quiver(0, 0, vetor2[0], vetor2[1], color='gray', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='Original v2')

    # Vetores originais (colocar por cima)
    plt.quiver(0, 0, vetor1[0], vetor1[1], color='gray', scale=0.8, scale_units='xy', angles='xy', label='Original v1', zorder=3)
    plt.quiver(0, 0, vetor2[0], vetor2[1], color='gray', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='Original v2', zorder=3)

    # Ap√≥s transforma√ß√£o
    plt.quiver(0, 0, v1_t[0], v1_t[1], color='blue', scale=1, scale_units='xy', angles='xy', label='A * v1')
    plt.quiver(0, 0, v2_t[0], v2_t[1], color='blue', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='A * v2')

    # Ap√≥s aplicar inversa
    plt.quiver(0, 0, v1_inv[0], v1_inv[1], color='red', scale=1, scale_units='xy', angles='xy', label='A‚Åª¬π * (A * v1)')
    plt.quiver(0, 0, v2_inv[0], v2_inv[1], color='red', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='A‚Åª¬π * (A * v2)')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.title("Transforma√ß√£o e Revers√£o pela Inversa")
    plt.legend()
    plt.show()

# Exemplo:
A = np.array([[2, 1], [5, 3]])
plotar_inversa(A)
```

### üìå Conclus√£o

* A matriz inversa **"desfaz"** a transforma√ß√£o.
* S√≥ matrizes com **determinante ‚â† 0** possuem inversa.
* A identidade $I$ √© o "zero da multiplica√ß√£o de matrizes": $A \cdot A^{-1} = I$

## üü¶ **Diagonaliza√ß√£o de Matrizes: Conceito Intuitivo**

Imagine que voc√™ tem uma matriz $A$ e quer entender como ela transforma o espa√ßo. Se voc√™ puder escrever $A$ de uma forma mais "simples", isso ajuda muito nos c√°lculos.

A **diagonaliza√ß√£o** √© justamente isso: transformar $A$ em uma matriz diagonal $D$, que √© super f√°cil de trabalhar, especialmente para pot√™ncias de matrizes ou exponenciais.

### üìò **Defini√ß√£o**

Uma matriz $A$ √© **diagonaliz√°vel** se existe uma matriz $P$ invert√≠vel tal que:

$$A = P D P^{-1}$$

Onde:
* $D$ √© uma matriz **diagonal** (com os autovalores de $A$ na diagonal)
* As colunas de $P$ s√£o os **autovetores** de $A$

### ‚ú≥Ô∏è **Por que isso √© √∫til?**

Com a matriz diagonal $D$, voc√™ consegue fazer:

$$A^n = P D^n P^{-1}$$

E como $D^n$ √© muito f√°cil de calcular (basta elevar os elementos da diagonal), isso **simplifica muito os c√°lculos**, especialmente em sistemas din√¢micos, equa√ß√µes diferenciais, e machine learning.


### üìä Visualiza√ß√£o com Exemplo Num√©rico
Vamos mostrar na pr√°tica como a diagonaliza√ß√£o age sobre os vetores ‚Äî antes e depois ‚Äî para ver a diferen√ßa no espa√ßo transformado.

```python
# Matriz A
A = np.array([[3, 1],
              [0, 2]])

# Autovalores e autovetores
autovalores, autovetores = np.linalg.eig(A)

# Construindo matriz diagonal e matriz P
D = np.diag(autovalores)
P = autovetores
P_inv = np.linalg.inv(P)

# Verifica√ß√£o da diagonaliza√ß√£o: A = P D P‚Åª¬π
A_reconstruida = P @ D @ P_inv
print("Matriz A original:")
print(A)
print("Reconstru√≠da via diagonaliza√ß√£o:")
print(A_reconstruida)

# Vetores base para visualiza√ß√£o
vetores = np.array([[1, 0], [0, 1], [1, 1], [-1, 0.5]])

# Plotando transforma√ß√£o por A
plt.figure(figsize=(12, 5))

# Original
plt.subplot(1, 2, 1)
for v in vetores:
    v_t = A @ v
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    plt.quiver(0, 0, v_t[0], v_t[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.title("Transforma√ß√£o com A")
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Transforma√ß√£o com D (base diagonal)
plt.subplot(1, 2, 2)
for v in vetores:
    v_diag = P_inv @ v
    v_t = D @ v_diag
    v_final = P @ v_t
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    plt.quiver(0, 0, v_final[0], v_final[1], angles='xy', scale_units='xy', scale=1, color='orange')
plt.title("Transforma√ß√£o com P D P‚Åª¬π")
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### ‚úÖ **Resumo da Diagonaliza√ß√£o**

* **Diagonalizar** uma matriz √© uma forma de **reescrev√™-la** com base nos seus autovalores e autovetores.
* A f√≥rmula central √©:

$$A = P D P^{-1}$$

Onde:
   * $D$ tem os autovalores na diagonal.
   * As colunas de $P$ s√£o os autovetores.

* A diagonaliza√ß√£o **simplifica** opera√ß√µes como eleva√ß√£o √† pot√™ncia:

$$A^n = P D^n P^{-1}$$

* Na pr√°tica, usamos isso em:
   * Sistemas din√¢micos lineares
   * Processos iterativos (ex: PageRank, Markov)
   * Machine Learning avan√ßado (ex: PCA, decomposi√ß√µes)

### üìå Considera√ß√µes finais

Nem toda matriz √© diagonaliz√°vel, mas quando √©, vale muito a pena us√°-la nessa forma para otimizar c√°lculos e revelar a estrutura interna da transforma√ß√£o.
