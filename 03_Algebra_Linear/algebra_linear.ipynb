### 2. Multiplicação de Matrizes

A multiplicação de matrizes é uma operação central na Álgebra Linear. Em IA, ela aparece em quase todos os cálculos envolvendo redes neurais, transformações lineares e representação de dados.

#### Regras principais:
Para multiplicar duas matrizes:
- A matriz **A** deve ter dimensões `m × n`
- A matriz **B** deve ter dimensões `n × p`
- O resultado será uma nova matriz **C** com dimensões `m × p`

Ou seja, o **número de colunas da primeira matriz deve ser igual ao número de linhas da segunda**.

#### Exemplo prático
Vamos visualizar a multiplicação de duas matrizes simples:


### 💡 Multiplicação de Matrizes: Exemplo 2x3 ⋅ 3x2

Se temos:

**Matriz A (2×3):**

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}
$$

**Matriz B (3×2):**

$$
B =
\begin{bmatrix}
7 & 8 \\
9 & 10 \\
11 & 12
\end{bmatrix}
$$

Então o produto \( C = A \cdot B \) será uma **matriz 2×2**, calculada assim:

$$
C = 
\begin{bmatrix}
(1\cdot7 + 2\cdot9 + 3\cdot11) & (1\cdot8 + 2\cdot10 + 3\cdot12) \\
(4\cdot7 + 5\cdot9 + 6\cdot11) & (4\cdot8 + 5\cdot10 + 6\cdot12)
\end{bmatrix}
$$

Fazendo as contas:

- \( C_{11} = 7 + 18 + 33 = 58 \)
- \( C_{12} = 8 + 20 + 36 = 64 \)
- \( C_{21} = 28 + 45 + 66 = 139 \)
- \( C_{22} = 32 + 50 + 72 = 154 \)

**Resultado final:**

$$
C = 
\begin{bmatrix}
58 & 64 \\
139 & 154
\end{bmatrix}
$$


```python
# Matriz A (2x3)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Matriz B (3x2)
B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Multiplicação
C = A @ B
print("Resultado da multiplicação:")
print(C)
```

#### Interpretação
Cada elemento da matriz resultado é o produto escalar entre uma linha da primeira matriz e uma coluna da segunda matriz.

Essa operação aparece diretamente em redes neurais, quando multiplicamos as entradas de um neurônio pelos seus respectivos pesos para obter uma ativação.


### 💡 Produto de Matrizes – Fórmula

Queremos multiplicar duas matrizes \( A \) e \( B \).  
Se:

$$

$$

a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22}
\end{bmatrix}
$$

### 📘 Multiplicação de Matrizes

Considere as matrizes:

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
\end{bmatrix},
\quad
B =
\begin{bmatrix}
7 & 8 \\
9 & 10 \\
11 & 12 \\
\end{bmatrix}
$$

Multiplicando \( A \cdot B \), obtemos a matriz \( C \):

$$
C =
\begin{bmatrix}
1 \cdot 7 + 2 \cdot 9 + 3 \cdot 11 & 1 \cdot 8 + 2 \cdot 10 + 3 \cdot 12 \\
4 \cdot 7 + 5 \cdot 9 + 6 \cdot 11 & 4 \cdot 8 + 5 \cdot 10 + 6 \cdot 12 \\

=
\begin{bmatrix}

```python
# Matrizes
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

# Multiplicação
C = A @ B
print("Resultado da multiplicação (C = A @ B):")
print(C)

# Visualização
fig, ax = plt.subplots(figsize=(6, 6))

# Vetores da base padrão
base = np.array([[1, 0], [0, 1]])

# Aplicando transformação da matriz C
for v in base:
    v_transformado = C @ v
    ax.quiver(0, 0, v[0], v[1], color='blue', angles='xy', scale_units='xy', scale=1, label='Base original' if np.array_equal(v, [1, 0]) else "")
    ax.quiver(0, 0, v_transformado[0], v_transformado[1], color='red', angles='xy', scale_units='xy', scale=1, label='Transformado' if np.array_equal(v, [1, 0]) else "")

ax.set_xlim(-10, 160)
ax.set_ylim(-10, 160)
ax.set_aspect('equal')
ax.axhline(0, color='gray')
ax.axvline(0, color='gray')
ax.grid(True)
ax.legend()
ax.set_title("Transformação Linear pela Matriz C = A @ B")
plt.show()
```

### 🧠 Produto Escalar – Intuição e Conceito
O produto escalar (ou produto interno) entre dois vetores é uma operação que gera um número (escalar), e é uma das formas de medir a relação entre eles.

Antes de expandirmos para operações mais avançadas com matrizes, é importante entender um conceito fundamental: o **produto escalar**.

## 🌟 O que é o Produto Escalar?

O produto escalar (ou produto interno) entre dois vetores resulta em um **número escalar** (não um vetor). Ele nos ajuda a entender:
* Quão "alinhados" dois vetores estão
* O ângulo entre eles
* A projeção de um vetor sobre outro

## 📐 Fórmula Matemática

Se temos dois vetores:

$$\vec{u} = [u_1, u_2, \dots, u_n], \quad \vec{v} = [v_1, v_2, \dots, v_n]$$

O produto escalar é:

$$\vec{u} \cdot \vec{v} = u_1v_1 + u_2v_2 + \dots + u_nv_n$$

## 🧮 Exemplo Numérico

Seja:

$$\vec{u} = [1, 2, 3], \quad \vec{v} = [4, 5, 6]$$

$$\vec{u} \cdot \vec{v} = 1 \cdot 4 + 2 \cdot 5 + 3 \cdot 6 = 4 + 10 + 18 = 32$$

## 🔍 Interpretação Geométrica

O produto escalar também pode ser visto como:

$$\vec{u} \cdot \vec{v} = \|\vec{u}\| \cdot \|\vec{v}\| \cdot \cos(\theta)$$

Onde:
* $\|\vec{u}\|$ e $\|\vec{v}\|$ são os módulos dos vetores
* $\theta$ é o **ângulo entre os vetores**

## 📐 Interpretação Geométrica

O produto escalar mede **quanto um vetor "aponta" na direção do outro**. Algumas consequências:

* Se o ângulo $\theta = 0°$ → $\cos(\theta) = 1$: vetores apontam na **mesma direção**, e o escalar é **máximo**.
* Se $\theta = 90°$ → $\cos(\theta) = 0$: vetores são **ortogonais (perpendiculares)**, e o produto escalar é **zero**.
* Se $\theta = 180°$ → $\cos(\theta) = -1$: vetores em **direções opostas**, e o escalar é **negativo**.

### 🧮 Exemplo Numérico

```python
u = np.array([3, 4])
v = np.array([2, -1])

produto_escalar = np.dot(u, v)
print("Produto escalar:", produto_escalar)
```

### 📊 Visualização

```python
def plot_vectors(u, v):
    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', label='v')
    plt.xlim(-1, 5)
    plt.ylim(-3, 5)
    plt.axhline(0, color='gray')
    plt.axvline(0, color='gray')
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Visualização dos vetores u e v")
    plt.show()

plot_vectors(u, v)
```

### ✅ Conclusões Importantes
Produto escalar é uma ferramenta essencial para identificar ângulos e relações entre vetores.

Se o produto escalar for zero, os vetores são perpendiculares.

Ele será positivo quando vetores estiverem "na mesma direção", e negativo quando "em direções opostas".

## 🔲 Matrizes Identidade e Inversa

### 🧱 Matriz Identidade

A matriz identidade $I$ é o **"1" da multiplicação de matrizes**. Ou seja:

$$A \cdot I = I \cdot A = A$$

Para uma matriz 2x2:

$$I = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$$

Ela **não altera** a matriz com a qual é multiplicada.

### 🔄 Matriz Inversa

A matriz inversa $A^{-1}$ é aquela que, quando multiplicada por $A$, resulta na matriz identidade:

$$A \cdot A^{-1} = A^{-1} \cdot A = I$$


### 🧮 Exemplo em Python

```python
A = np.array([[2, 1],
              [5, 3]])

# Verificando a identidade
I = np.eye(2)

# Inversa de A
A_inv = np.linalg.inv(A)

# Testando
produto = A @ A_inv

print("Matriz A:")
print(A)

print("\nInversa de A:")
print(A_inv)

print("\nA @ A_inv (deve ser identidade):")
print(produto)
```

### 📐 Geometria da Inversa
Visualmente, aplicar uma matriz e depois sua inversa é como voltar ao ponto original.

Por exemplo: uma transformação que gira um vetor em 90° e depois aplica a inversa, desfaz a rotação.

### ⚠️ Determinante

Para saber se uma matriz tem inversa, usamos o **determinante**.

* Se $\text{det}(A) \neq 0$: matriz invertível
* Se $\text{det}(A) = 0$: **matriz singular** (sem inversa)

### 🧮 Exemplo em Python

```python
np.linalg.det(A)
```

### ⛔ Casos sem Inversa

```python
B = np.array([[1, 2],
              [2, 4]])

print("Determinante de B:", np.linalg.det(B))
```

Resultado: 0 → A matriz é singular (inversão proibida).

## 📊 Sistemas Lineares e Solução com Matrizes

### 🧩 O Problema

Imagine um sistema de equações lineares:

$$\begin{cases} 
2x + y = 5 \\ 
4x - 3y = 6 
\end{cases}$$

Esse sistema pode ser representado como **uma equação matricial**:

$$A \cdot \vec{x} = \vec{b}$$

Onde:
* $A = \begin{bmatrix} 2 & 1 \\ 4 & -3 \end{bmatrix}$
* $\vec{x} = \begin{bmatrix} x \\ y \end{bmatrix}$
* $\vec{b} = \begin{bmatrix} 5 \\ 6 \end{bmatrix}$

### 🎯 Objetivo

Resolver o sistema significa encontrar $\vec{x}$, ou seja:

$$\vec{x} = A^{-1} \cdot \vec{b}$$

### 🧮 Solução em Python

```python
# Matriz dos coeficientes
A = np.array([[2, 1],
              [4, -3]])

# Vetor dos resultados
b = np.array([5, 6])

# Solução do sistema
x = np.linalg.solve(A, b)

print("Solução do sistema (valores de x e y):")
print(x)
```

### 📉 Interpretação Geométrica
Cada equação representa uma reta no plano. A solução do sistema é o ponto de interseção dessas duas retas.

```python
x_vals = np.linspace(-10, 10, 100)
y1 = (5 - 2 * x_vals)
y2 = (4 * x_vals - 6) / 3

plt.plot(x_vals, y1, label='2x + y = 5')
plt.plot(x_vals, y2, label='4x - 3y = 6')
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.title('Solução de um Sistema Linear: Interseção das retas')
plt.show()
```

### 🚧 Quando não há solução?

Existem 3 possibilidades para um sistema:

| Tipo de sistema | Situação geométrica | Solução |
|-----------------|---------------------|---------|
| Determinado | Retas se cruzam | 1 solução (interseção) |
| Indeterminado | Retas coincidem | ∞ soluções |
| Impossível | Retas paralelas | Nenhuma solução |

## 🔍 Autovalores e Autovetores: Conceito Intuitivo

### 🧠 Ideia Principal

Imagine que uma transformação linear (representada por uma matriz $A$) muda a direção da maioria dos vetores. Mas existem vetores **especiais** que **mantêm a sua direção** após a transformação — apenas mudam de tamanho.

Esses vetores são chamados **autovetores**, e o fator pelo qual são multiplicados é o **autovalor**.

### 🔎 Definição Matemática

Dado um vetor $\vec{v} \neq \vec{0}$ e uma matriz quadrada $A$, dizemos que:

$$A \vec{v} = \lambda \vec{v}$$

Se:
* $\vec{v}$: autovetor
* $\lambda$: autovalor correspondente

Isso significa: **multiplicar $A$ pelo vetor $\vec{v}$ é o mesmo que multiplicar esse vetor por um escalar $\lambda$**.

### 🧮 Exemplo em Python

```python
# Matriz de transformação
A = np.array([
    [2, 1],
    [1, 2]
])

# Cálculo de autovalores e autovetores
valores, vetores = np.linalg.eig(A)

print("Autovalores:")
print(valores)

print("\nAutovetores (colunas):")
print(vetores)
```

### 🎯 Interpretação Geométrica
Vamos visualizar como autovetores são transformados pela matriz:

```python
# Pega os dois autovetores
v1 = vetores[:, 0]
v2 = vetores[:, 1]

# Aplica a transformação
Av1 = A @ v1
Av2 = A @ v2

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
plt.quiver(0, 0, Av1[0], Av1[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='A*v1')

#plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')
#plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=1, color='orange', label='A*v2')

#plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='v2')
#plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=0.8, color='orange', alpha=0.8, label='A*v2')


plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, label='v2')
plt.quiver(0, 0, Av2[0], Av2[1], angles='xy', scale_units='xy', scale=0.8, color='orange', alpha=0.9, label='A*v2')
plt.legend()


plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Autovetores mantêm a direção após transformação")
plt.show()
```

🔷 Determinante de uma Matriz – Significado Geométrico e Intuição
O determinante de uma matriz quadrada representa, geometricamente, o fator de escala da área (em 2D) ou do volume (em 3D) após uma transformação linear.

✅ Intuição Visual:
- Se a matriz representa uma transformação no plano (2D), o determinante indica quanto a área é multiplicada.

- Se o determinante for zero, significa que a transformação achata o espaço, ou seja, colapsa o plano em uma linha (ou ponto).

- Se for negativo, além da escala, há uma inversão de orientação (reflexão).

### 🔢 Código Python: Visualização do Efeito do Determinante

```python
def plotar_determinante(matriz, titulo=""):
    # Define o quadrado unitário
    quadrado = np.array([
        [0, 1, 1, 0, 0],
        [0, 0, 1, 1, 0]
    ])

    # Aplica a transformação
    transformado = matriz @ quadrado

    # Calcula o determinante
    det = np.linalg.det(matriz)

    # Plot
    plt.figure(figsize=(6, 6))
    plt.plot(quadrado[0], quadrado[1], 'gray', label='Original')
    plt.plot(transformado[0], transformado[1], 'r', label='Transformado')
    plt.fill(transformado[0], transformado[1], 'r', alpha=0.2)

    plt.title(f"{titulo}\nDeterminante: {det:.2f}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.show()

# Exemplos:
plotar_determinante(np.array([[2, 0], [0, 1]]), "Escala em X (aumenta a área)")
plotar_determinante(np.array([[1, 0], [0, -1]]), "Reflexão no eixo Y (inverte a orientação)")
plotar_determinante(np.array([[1, 1], [1, 1]]), "Cisalhamento colapsante (det = 0)")
```

### 📌 Conclusão:
Determinante = 1 → a área não muda.

Determinante > 1 → a área aumenta.

Determinante < 0 → a área muda e há inversão de sentido (reflexão).

Determinante = 0 → colapso de dimensão, transformação não-inversível.

## 🔷 Matriz Inversa – Intuição, Propriedades e Visualização

A **matriz inversa** é o equivalente de "desfazer" uma transformação linear.

Se uma matriz **A** transforma vetores, a **inversa de A**, denotada $A^{-1}$, **desfaz** essa transformação.

### ✅ Intuição:

Se aplicarmos uma matriz $A$ e depois sua inversa $A^{-1}$, voltamos ao ponto de origem:

$$A^{-1} \cdot (A \cdot \vec{v}) = \vec{v}$$

Ou seja, **transformar e depois reverter**.

### 🧠 Condição:

Uma matriz **só tem inversa se seu determinante for diferente de zero**:

$$\text{Se } \det(A) \neq 0 \Rightarrow A \text{ é invertível}$$

### 🔢 Exemplo Numérico:
Vamos ver um exemplo simples de inversão de matriz 2×2.

```python
# Matriz A
A = np.array([
    [2, 1],
    [5, 3]
])

# Inversa de A
A_inv = np.linalg.inv(A)

# Produto A . A⁻¹ (deve dar a identidade)
I = A @ A_inv

print("Matriz A:")
print(A)

print("\nMatriz Inversa de A:")
print(A_inv)

print("\nA . A⁻¹ (deve ser a identidade):")
print(I)
```

### 🎯 Visualização: Transformar e "Desfazer"

```python
def plotar_inversa(A):
    A_inv = np.linalg.inv(A)

    # Vetores originais
    vetor1 = np.array([1, 1])
    vetor2 = np.array([-1, 1])

    # Aplica a transformação A
    v1_t = A @ vetor1
    v2_t = A @ vetor2

    # Aplica a inversa em seguida
    v1_inv = A_inv @ v1_t
    v2_inv = A_inv @ v2_t

    plt.figure(figsize=(6, 6))

    # Vetores originais
    #plt.quiver(0, 0, vetor1[0], vetor1[1], color='gray', scale=1, scale_units='xy', angles='xy', label='Original v1')
    #plt.quiver(0, 0, vetor2[0], vetor2[1], color='gray', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='Original v2')

    # Vetores originais (colocar por cima)
    plt.quiver(0, 0, vetor1[0], vetor1[1], color='gray', scale=0.8, scale_units='xy', angles='xy', label='Original v1', zorder=3)
    plt.quiver(0, 0, vetor2[0], vetor2[1], color='gray', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='Original v2', zorder=3)

    # Após transformação
    plt.quiver(0, 0, v1_t[0], v1_t[1], color='blue', scale=1, scale_units='xy', angles='xy', label='A * v1')
    plt.quiver(0, 0, v2_t[0], v2_t[1], color='blue', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='A * v2')

    # Após aplicar inversa
    plt.quiver(0, 0, v1_inv[0], v1_inv[1], color='red', scale=1, scale_units='xy', angles='xy', label='A⁻¹ * (A * v1)')
    plt.quiver(0, 0, v2_inv[0], v2_inv[1], color='red', alpha=0.5, scale=1, scale_units='xy', angles='xy', label='A⁻¹ * (A * v2)')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.xlim(-6, 6)
    plt.ylim(-6, 6)
    plt.title("Transformação e Reversão pela Inversa")
    plt.legend()
    plt.show()

# Exemplo:
A = np.array([[2, 1], [5, 3]])
plotar_inversa(A)
```

### 📌 Conclusão

* A matriz inversa **"desfaz"** a transformação.
* Só matrizes com **determinante ≠ 0** possuem inversa.
* A identidade $I$ é o "zero da multiplicação de matrizes": $A \cdot A^{-1} = I$

## 🟦 **Diagonalização de Matrizes: Conceito Intuitivo**

Imagine que você tem uma matriz $A$ e quer entender como ela transforma o espaço. Se você puder escrever $A$ de uma forma mais "simples", isso ajuda muito nos cálculos.

A **diagonalização** é justamente isso: transformar $A$ em uma matriz diagonal $D$, que é super fácil de trabalhar, especialmente para potências de matrizes ou exponenciais.

### 📘 **Definição**

Uma matriz $A$ é **diagonalizável** se existe uma matriz $P$ invertível tal que:

$$A = P D P^{-1}$$

Onde:
* $D$ é uma matriz **diagonal** (com os autovalores de $A$ na diagonal)
* As colunas de $P$ são os **autovetores** de $A$

### ✳️ **Por que isso é útil?**

Com a matriz diagonal $D$, você consegue fazer:

$$A^n = P D^n P^{-1}$$

E como $D^n$ é muito fácil de calcular (basta elevar os elementos da diagonal), isso **simplifica muito os cálculos**, especialmente em sistemas dinâmicos, equações diferenciais, e machine learning.


### 📊 Visualização com Exemplo Numérico
Vamos mostrar na prática como a diagonalização age sobre os vetores — antes e depois — para ver a diferença no espaço transformado.

```python
# Matriz A
A = np.array([[3, 1],
              [0, 2]])

# Autovalores e autovetores
autovalores, autovetores = np.linalg.eig(A)

# Construindo matriz diagonal e matriz P
D = np.diag(autovalores)
P = autovetores
P_inv = np.linalg.inv(P)

# Verificação da diagonalização: A = P D P⁻¹
A_reconstruida = P @ D @ P_inv
print("Matriz A original:")
print(A)
print("Reconstruída via diagonalização:")
print(A_reconstruida)

# Vetores base para visualização
vetores = np.array([[1, 0], [0, 1], [1, 1], [-1, 0.5]])

# Plotando transformação por A
plt.figure(figsize=(12, 5))

# Original
plt.subplot(1, 2, 1)
for v in vetores:
    v_t = A @ v
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    plt.quiver(0, 0, v_t[0], v_t[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.title("Transformação com A")
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Transformação com D (base diagonal)
plt.subplot(1, 2, 2)
for v in vetores:
    v_diag = P_inv @ v
    v_t = D @ v_diag
    v_final = P @ v_t
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    plt.quiver(0, 0, v_final[0], v_final[1], angles='xy', scale_units='xy', scale=1, color='orange')
plt.title("Transformação com P D P⁻¹")
plt.axhline(0, color='k', linewidth=0.5)
plt.axvline(0, color='k', linewidth=0.5)
plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

### ✅ **Resumo da Diagonalização**

* **Diagonalizar** uma matriz é uma forma de **reescrevê-la** com base nos seus autovalores e autovetores.
* A fórmula central é:

$$A = P D P^{-1}$$

Onde:
   * $D$ tem os autovalores na diagonal.
   * As colunas de $P$ são os autovetores.

* A diagonalização **simplifica** operações como elevação à potência:

$$A^n = P D^n P^{-1}$$

* Na prática, usamos isso em:
   * Sistemas dinâmicos lineares
   * Processos iterativos (ex: PageRank, Markov)
   * Machine Learning avançado (ex: PCA, decomposições)

### 📌 Considerações finais

Nem toda matriz é diagonalizável, mas quando é, vale muito a pena usá-la nessa forma para otimizar cálculos e revelar a estrutura interna da transformação.
